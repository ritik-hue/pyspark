```python
from pyspark.sql import SparkSession
spark=SparkSession.builder.appName('Practise').getOrCreate()
```

```python
df_pyspark=spark.read.csv('/content/drive/MyDrive/Weather_Data.csv',header=True,inferSchema=True)
```

```python
df_pyspark.show()
```

```python
df_pyspark.drop('Evaporation').show()
```

```python
df_pyspark.na.drop().show()

```

```python
##drop any column that have null value
df_pyspark.na.drop(how="any").show()
```

```python
##if there is 3 non null values(threshold) its gonna keep it else its gonna delete the row
df_pyspark.na.drop(how="any",thresh=3).show()
```

```python
df_pyspark.na.drop(how="any",subset=['Rainfall']).show()
```

```python
df_pyspark.na.fill('Missing Values',['Rainfall','Evaporation']).show()
```

```python
##replacing missing values with mean
from pyspark.ml.feature import Imputer
imputer = Imputer(
    inputCols=['age', 'Experience', 'Salary'],
    outputCols=["{}_imputed".format(c) for c in ['age', 'Experience', 'Salary']]
    ).setStrategy("median")
```

```python
# Add imputation cols to df
imputer.fit(df_pyspark).transform(df_pyspark).show()
```
